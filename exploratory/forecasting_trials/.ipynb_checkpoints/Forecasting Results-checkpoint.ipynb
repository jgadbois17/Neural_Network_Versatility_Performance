```python
import sys 
sys.path.append('../../') 

import numpy as np 
import pandas as pd 
import tensorflow as tf 
import matplotlib.pyplot as plt 

from dspML import data, plot 
from dspML.preprocessing import sequence 
from dspML.models.sequence import arima, fc_rnn, fc_cnn 
from dspML.evaluation import ForecastEval 
```

<a name="1"></a> 

# 1. Load and Plot Time Series 



```python
# data paths 
path_train = '../../dspML/datasets/daily_climate/DailyDelhiClimateTrain.csv' 
path_test = '../../dspML/datasets/daily_climate/DailyDelhiClimateTest.csv' 

# load full dataset 
df = pd.read_csv(path_train) 
df = df.append(pd.read_csv(path_test)) 
df['date'] = pd.to_datetime(df['date'], yearfirst=True) 
df.set_index('date', inplace=True) 
df.drop(['meantemp', 'meanpressure'], axis=1, inplace=True) 
print('Data Dimensions: ', df.shape) 
df.head() 
```

## Extract Humidity and Visualize 

```python
# humidity 
yh = df['humidity'].resample('D').mean() 

# plot signal 
plot.signal_pd(yh, title='Daily Humidity Time Series Signal', figsize=(16, 7)) 
```

```python
# plot ACF/PACF 
plot.p_acf(yh, lags=40) 
```

## Extract Wind Speed and Visualize 

```python
# extract wind speed 
yw = df['wind_speed'].resample('D').mean() 

# plot wind speed 
plot.signal_pd(yw, title='Daily Wind Speed Time Series Signal', figsize=(16, 7)) 
```

```python
# plot ACF/PACF 
plot.p_acf(yw, lags=40) 
```

# 2. Forecasting with ARIMA 

* [Outline](#0) 

---

## Forecast Humidity 

```python
# split data 
fc_hzn = 7 
yh_train, yh_test = sequence.temporal_split(yh, fc_hzn) 

# define and fit model 
model = arima.ARIMA(yh_train, order=(1,1,1), seasonal_order=(0,1,1,12)) 
```

```python
# predict forecast and evaluate 
yh_pred = model.forecast(steps=7) 
fc_eval = ForecastEval(yh_test, yh_pred) 
fc_eval.mse() 
```

```python
# plot forecast 
plot.time_series_forecast(series=yh_train.iloc[-50:], 
                          y=yh_test, 
                          y_pred=yh_pred, 
                          title='Humidity 7-Day Forecast', 
                          figsize=(16, 7)) 
```
